```python
def h(X , theta):
          # print(np.dot(X, theta))
    return np.dot(X, theta)
```

```python
def cout(X,theta , y):
          return (1/2)*np.sum((h(X , theta) - y)**2)
```

```python
def gradient_descent(X, theta , y , pas):
          return theta - (pas *  (np.dot(X.T , (h(X , theta) - y)))) 
```

```python
def fit(X , theta , y , pas , nb_iter):
    for i in range(nb_iter):
        theta = gradient_descent(X , theta , y , pas)
    return theta


theta = fit(X , theta , y , pas , 10000)
theta
```
