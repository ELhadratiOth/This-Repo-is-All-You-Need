```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
img = cv2.imread('img3.jpg')

plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Convert to RGB 
# cv2.imshow('Image Window', img) # Display the image in a siparate window
# cv2.waitKey(0)
# cv2.destroyAllWindows() 

plt.imshow(img)
img.shape

# the 400*612 is the size of the image
# the  ,3 is the number of channels in the image (RGB)
```

```python
# getting the  r g b values of the image
r , g , b = cv2.split(img)

# scalling the values of the image
r = r/255.0
g = g/255.0
b = b/255.0

plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
pca_components = 50 # number of components to keep in the PCA from 612 columns of the image data to 50 columns

pca_r = PCA(n_components=pca_components)
pca_g = PCA(n_components=pca_components)
pca_b = PCA(n_components=pca_components)
reduc_r = pca_r.fit_transform(r)
reduc_g = pca_g.fit_transform(g)
reduc_b = pca_b.fit_transform(b)

# here we are reducing the number of columns from 612 to 50 columns for  each of the r g b values of the image (channels)
 
```

```python
reconstructed_r = pca_r.inverse_transform(reduc_r)
reconstructed_g = pca_g.inverse_transform(reduc_g)
reconstructed_b = pca_b.inverse_transform(reduc_b) 
# after reducing the number of columns we are now reconstructing the image with the reduced columns but the quality of the image will be reduced 
img_reconstructed = np.stack([reconstructed_r, reconstructed_g, reconstructed_b], axis=2)

plt.imshow(img_reconstructed)
```
